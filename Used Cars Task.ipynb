$\text{Margin Profit} = \left( \frac{\text{Revenue} - \text{Cost}}{\text{Revenue}} \right) \times 100$


```python
df[["Price-$","Sold Price-$"]].head(3)
```

```python
print(df["Price-$"].dtype)
```

```python
df["Margin Profit"]= (df["Sold Price-$"]-df["Price-$"])/100
```

<font size=6 color=yellow>



- Margin Profit

```python
df['Margin Profit']
```

<font size=3 color=yellow>

$\text{Final Price} = Price - \left( Price \times \frac{\text{Margin Profit}}{100} \right) - Commission$


```python
df["Final Price"] = (df["Price-$"]-(df["Price-$"]*(df["Margin Profit"]/100))-df["Sales Commission-$"])
```

## Q7 ✅
* Generate pivot tables for indexing by the following categories: (Manufacturer Name, Car Type, Gearbox, Feedback). The values will represent the (mean/min/max) sold price for each category. and the same for Purchased Price

* save each of them in a separate sheet within an Excel file 

```python
manufacturer_name = df.pivot_table(aggfunc=['mean','min','max'],index="Manufacturer Name",values='Sold Price-$',)
print(manufacturer_name)
```

```python
car_type = df.pivot_table(aggfunc=['mean','min','max'],index="Car Type",values='Sold Price-$',)
print(car_type)
```

```python
gearbox = df.pivot_table(aggfunc=['mean','min','max'],index="Gearbox",values='Sold Price-$',)
print(gearbox)
```

```python
feedback= df.pivot_table(aggfunc=['mean','min','max'],index="Feedback",values='Sold Price-$',)
print(feedback)
```

```python
pivots=[manufacturer_name,car_type,gearbox,feedback]
```

```python
with pd.ExcelWriter("./output/Analysis data.xlsx") as file:
    df.to_excel(file ,sheet_name="Main table", index=False)
    for table in pivots:
        table.to_excel(file, sheet_name=f"pivot_{table.index.name}", index=True)
```
